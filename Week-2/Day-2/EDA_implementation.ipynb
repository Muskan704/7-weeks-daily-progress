```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

DATA CLEANING

```python
# missing values
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Remove the null values
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# Fix spaces in the columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
# Change the required col as integer datatype
df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
# Changing the other col to float datatype
object = [features for features in df.columns if df[features].dtypes=='string']   # Always remember the type of the data you are converting and then write the correct dtype. 
```

```python
for i in object:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
# Let's save the clean dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
